```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
df = pd.read_csv('Graduation_Dropout.csv')
```

```python
df.head()
```

```python
df = df[df.Subgroup == 'All Students']
df = df[df.BuildingName != 'State']
df = df[df.BuildingName != '00000']
df = df[df.RateYear == '4-Year']
df = df[df.GraduationRate != '0.00']
df = df.assign(BuildingName_clean=df.BuildingName.str[:-8])
df_cleaned = df[['CountyName','BuildingName_clean','GraduationRate']]
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
for i in df_cleaned.index:
    name = df_cleaned.loc[i, 'BuildingName_clean']
    county = df_cleaned.loc[i, 'CountyName']
    print(f'{name}, {county} County, MI')
```

```python
for i in df_cleaned.index:
    name = df_cleaned.loc[i, 'BuildingName_clean']
    county = df_cleaned.loc[i, 'CountyName']
    query = f'{name}, {county} County, MI'
    r = requests.get(f'https://maps.googleapis.com/maps/api/place/textsearch/json?query={query}&key=AIzaSyD5NFzc0pfFbJyCPPz-SqMdnLYuIGNpDE8')
    res = r.json()
    if res['results']:
        df_cleaned.loc[i, 'latitude'] = str(res['results'][0]['geometry']['location']['lat'])
        df_cleaned.loc[i, 'longitude'] = str(res['results'][0]['geometry']['location']['lng'])
    else:
        df_cleaned.loc[i, 'latitude'] = '0'
        df_cleaned.loc[i, 'longitude'] = '0'
```
